### Dungeon Egg Hunt

Using object-oriented programming principles, create an egg hunt game. Your objective is to go through the dungeon to collect the basket and the eggs. Once you've retrieved all of the items, exit through the door.

**RULES**:

`1. ` Player, Monster, Door, Basket and the 3 Eggs must all start at a random locations on a grid.

`2. ` Player must collect the Basket first before collecting the 3 Eggs. Player will not be able to collect any Eggs until it first has collected the Basket.

`3. ` Once Player collects the Basket, then and only then will it be able to collect any Eggs.

`4. ` When Player collects the Basket and the 3 Eggs, Player will need to head toward the Door.

`5. ` The entire time the Player is searching for the Eggs, Basket and the Door, there will be a Monster trying to eat the Player.

`6. ` If the Monster lands on the same space as the Player, or the Player lands on the same space as the Monster, the game ends and the game will prompt the user to play again.

`7. ` If Player reaches the Door before the Monster catches them, you win the game and the game will prompt the user to play again.

`8. ` The Player does not know where any of the other tokens are, so all of the tokens with the exception of the Player will be hidden.

**Note**: For debugging purposes, you can create tokens for each of the other tokens. For example:<br>
* Player (P)
* Monster (M)
* Eggs (O)
* Basket (U)
* Door (D)

```python
import random
from IPython.display import clear_output

CELLS = [
    (0,0),(1,0),(2,0),(3,0),(4,0),
    (0,1),(1,1),(2,1),(3,1),(4,1),
    (0,2),(1,2),(2,2),(3,2),(4,2),
    (0,3),(1,3),(2,3),(3,3),(4,3),
    (0,4),(1,4),(2,4),(3,4),(4,4)
      ]
def instructions():
    print("""Welcome to the dungeon! The object of this game is to find the basket, 3 eggs, and then the door.
You are not alone..... there is a monster here, hungry and waiting. But here's the catch. You don't know where he is.
Good luck!! """)
class Game:
    
    
    def __init__(self, cells, player, monster, door, basket, egg1, egg2, egg3):
        self.cells = cells
        self.player = player
        self.monster = monster
        self.door = door
        self.basket = basket
        self.egg1 = egg1
        self.egg2 = egg2
        self.egg3 = egg3
        
    
    def draw_map(self):
        count = 0
        for cell in self.cells:
            count += 1
            if cell == self.player.get_location():
                val = "X"
            elif cell == self.monster.get_location():
                val = "M"
            elif cell == self.egg1.get_location():
                val = "E"
            elif cell == self.egg2.get_location():
                val = "E"
            elif cell == self.egg3.get_location():
                val = "E" 
            elif cell == self.basket.get_location():
                val = "B"
            elif cell == self.door.get_location():
                val = "D"
            else:
                val = "_"
                
            string = f"|{val}|"
            if count == 5:
                string += "\n"
                count = 0
            print(string, end = '')
            
    
                   
        
        
    
        
class Token:
    def __init__(self):
        self.location = CELLS[random.randint(0, len(CELLS) - 1)]
    
    def get_location(self):    
        return self.location
    
    def set_location(self, location):
        self.location = location
    
class Player(Token):
    eggs = []
    has_basket = False
    
    def has_eggs(self):
        return len(self.eggs) == 3
    
    def move_player(self, move):
        x,y = self.location
       
            
        if move == 'w' and y > 0:
            y -= 1

        if move == 's' and y < 4:
            y += 1

        if move == 'a' and x > 0:
            x -= 1

        if move == 'd' and x < 4:
            x += 1
        self.set_location((x, y))
        

class Monster(Token):
    pass 

class Egg(Token):
    pass
    
class Basket(Token):
    pass

class Door(Token):
    pass


    
    
def main():
   
    p1 = Player()
    m1 = Monster()
    d1 = Door()
    b1 = Basket()
    e1 = Egg()
    e2 = Egg()
    e3 = Egg()
    g = Game(CELLS, p1, m1, d1, b1, e1, e2, e3)
     
    playing = True
    
    
    instructions()
    while playing:
        
        
        g.draw_map()
        p1.move_player(input("w,s,a,d >>> "))
        clear_output()
        if p1.location == m1.location:
            print("*** You have been eaten by the monster! Good luck next time! ***")
            playing = False
            if input("Play again? (y/n): ").lower != "n":
                clear_output()
                main()
            else:
                break
        elif p1.location == b1.location:
            print("*** You've found the basket! Now you can collect the eggs! ***")
            p1.has_basket = True
        elif p1.location == e1.location or p1.location == e2.location or p1.location == e3.location:
            if not p1.has_basket:
                print("You must find the basket before you can collect eggs! ")
            else:
                if p1.location == e1.location:
                    print("*** You've found an egg! ***")
                    p1.eggs.append(e1)
                elif p1.location == e2.location:
                    print("*** You've found an egg! ***")
                    p1.eggs.append(e2)
                elif p1.location == e3.location:
                    print("*** You've found an egg! ***")
                    p1.eggs.append(e3)
        else:
            playing = True
        
        if p1.location == d1.location:
            if not p1.has_basket and not p1.has_eggs():
                print("You need the basket and have all 3 eggs to end game! ")
                playing = True
            else:
                print("You have won the game!")
                playing = False
                
                if input("Play again? (y/n): ").lower != "n":

                    clear_output()
                    main()
            
                

   

        
            
            
       
   
main() 
    
  
    
    
    
    
    
    
    
```
